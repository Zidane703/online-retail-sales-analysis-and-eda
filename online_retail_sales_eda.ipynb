# 1. Data Cleaning

```python
import pandas as pd

# Membaca dataset Online Retail dari Excel
df = pd.read_excel('online_retail_2.xlsx')

# Melihat 5 baris pertama data mentah
print("Data Mentah:")
print(df.head())
```

```python
# 1. Menghapus baris yang 'Customer ID'-nya kosong (Perhatikan ada SPASI)
df_clean = df.dropna(subset=['Customer ID']).copy()

# 2. Filter hanya pesanan yang sukses (Quantity > 0 dan Price > 0)
df_clean = df_clean[(df_clean['Quantity'] > 0) & (df_clean['Price'] > 0)]

# 3. Ubah tipe data 'Customer ID' menjadi String/Teks agar tidak dijumlahkan (SUM)
df_clean['Customer ID'] = df_clean['Customer ID'].astype(int).astype(str)

# 4. Ubah 'Invoice' dan 'StockCode' menjadi tipe teks
df_clean['Invoice'] = df_clean['Invoice'].astype(str)
df_clean['StockCode'] = df_clean['StockCode'].astype(str)

# 5. Buat kolom baru: Total Sales (Omzet = Quantity * Price)
df_clean['TotalSales'] = df_clean['Quantity'] * df_clean['Price']

print("Jumlah baris sebelum dibersihkan:", f"{len(df):,}")
print("Jumlah baris setelah dibersihkan :", f"{len(df_clean):,}")
```

```python
# 1. Menghapus baris yang 'Customer ID'-nya kosong (Missing Values)
df_clean = df.dropna(subset=['Customer ID']).copy()

# 2. Membuang transaksi retur/dibatalkan (Quantity minus atau Price <= 0)
df_clean = df_clean[(df_clean['Quantity'] > 0) & (df_clean['Price'] > 0)]

# 3. Mengubah 'Customer ID' dari desimal (13085.0) menjadi string/teks rapi ('13085')
df_clean['Customer ID'] = df_clean['Customer ID'].astype(int).astype(str)

# 4. Mengubah 'Invoice' dan 'StockCode' menjadi tipe teks (agar konsisten)
df_clean['Invoice'] = df_clean['Invoice'].astype(str)
df_clean['StockCode'] = df_clean['StockCode'].astype(str)

# 5. FEATURE ENGINEERING: Membuat kolom baru 'TotalSales' (Omzet)
df_clean['TotalSales'] = df_clean['Quantity'] * df_clean['Price']

# Menampilkan hasil pembersihan
print(f"Jumlah baris sebelum dibersihkan : {len(df):,}")
print(f"Jumlah baris setelah dibersihkan : {len(df_clean):,}")
print("-" * 50)
print("Data Bersih Siap untuk Project Setelahnya Mr. Zidane:")
print(df_clean[['Invoice', 'StockCode', 'Description', 'Quantity', 'Price', 'TotalSales', 'Customer ID', 'Country']].head())
```

# 4. Top 10 Revenue Contributing Countries & Top 10 Revenue Contributing Countries

```python
from IPython.display import display
import pandas as pd

# ==============================================================================
# 1. STATISTIK DESKRIPTIF YANG DIPERCANTIK
# ==============================================================================
print("📊 STATISTIK DESKRIPTIF DATA ONLINE RETAIL")
print("=" * 65)

# Membuat tabel statistik (kita ubah posisinya / Transpose dengan .T agar mudah dibaca ke samping)
stats_df = df_clean[['Quantity', 'Price', 'TotalSales']].describe().T

# Mengubah nama baris agar lebih deskriptif
stats_df.index = ['📦 Kuantitas (Quantity)', '🏷️ Harga Satuan (Price)', '💰 Omzet (TotalSales)']

# Menampilkan tabel dengan formatting rapi dan sorotan warna biru pada angka Mean & Max
display(stats_df.style.format({
    'count': '{:,.0f}',
    'mean': '{:,.2f}',
    'std': '{:,.2f}',
    'min': '{:,.2f}',
    '25%': '{:,.2f}',
    '50%': '{:,.2f}',
    '75%': '{:,.2f}',
    'max': '{:,.2f}'
}).background_gradient(cmap='Blues', subset=['mean', 'std', 'max']))


# ==============================================================================
# 2. PIVOT TABLE EKSEKUTIF YANG DIPERCANTIK
# ==============================================================================
print("\n🏆 TOP 10 NEGARA PENYUMBANG OMZET TERBESAR")
print("=" * 65)

# Membuat Pivot Table dengan multi-metrik (Total Omzet, Jumlah Item, dan Rata-rata)
pivot_negara = df_clean.pivot_table(
    index='Country', 
    values='TotalSales', 
    aggfunc=['sum', 'count', 'mean']
)

# Merapikan nama kolom
pivot_negara.columns = ['Total Omzet (£)', 'Jumlah Transaksi', 'Rata-rata per Item (£)']
pivot_negara = pivot_negara.sort_values(by='Total Omzet (£)', ascending=False).head(10)

# Menambahkan kolom persentase kontribusi omzet
total_semua_omzet = df_clean['TotalSales'].sum()
pivot_negara['Kontribusi (%)'] = (pivot_negara['Total Omzet (£)'] / total_semua_omzet) * 100

# Menata urutan kolom agar mudah dibaca
pivot_negara = pivot_negara[['Total Omzet (£)', 'Kontribusi (%)', 'Jumlah Transaksi', 'Rata-rata per Item (£)']]

# Menampilkan dengan format mata uang Poundsterling (£), persentase, dan Bar Grafik interaktif
display(pivot_negara.style.format({
    'Total Omzet (£)': '£{:,.2f}',
    'Kontribusi (%)': '{:.2f}%',
    'Jumlah Transaksi': '{:,.0f}',
    'Rata-rata per Item (£)': '£{:,.2f}'
}).background_gradient(cmap='Greens', subset=['Total Omzet (£)'])
  .bar(subset=['Kontribusi (%)'], color='#ff9999', vmin=0, vmax=100))
```
